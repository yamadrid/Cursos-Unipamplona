##Bucle While

El bucle while nos permite ejecutar un bloque de código continuamente mientras la condición sea verdadera e incluso podemos crear bucles infinitos.

Estos se ejecutan mientras la condición sea verdadera, es decir, si la condición se cumple se ejecutara el cuerpo de dicho bucle y al finalizar se volverá a comprobar la condición, si continua siendo verdadera pues se ejecutara nuevamente. En caso contrario omitirá el mismo y continuará la ejecución normal del programa. Venga veamos como crear un bucle while y para que nos ha de servir

```
# while (condición):
    Cuerpo de la repetición
```

1. Hacer la función Factorial

  $
  n!= \left\{ \begin{array}{lcc}
              0 &   si  & n = 0 \\
              \\ 1*2*3*4*....*n &  si & n > 0 \\
              \end{array}
    \right.
  $

````
def factorial(n):
    k = 1
    r = 1

    while (k<=n):
        r = r * k
        k = k + 1

    return r

factorial(5)
````

2. Dados dos (2) números ”x” y “z”donde “x” es menor que “z” haga una función para calcular la suma de los números comprendidos entre ”x” y “z”

````
def prob31(x,z):
    r = 1
    k = x + 1

    while (k<z):
        r = r * k
        k = k + 1

    return r

prob31(2,6)
````

3. Juan Pablo Montoya, entrena todos los días, recorriendo cierta cantidad de kilómetros, de tal forma que todos los días recorre el doble de lo que recorrió el día anterior más 10 km. Pero cada 3 días recorre solo la mitad de lo que recorrió el día anterior. Realice una función que dado el número de kilómetros recorridos el primer día del entrenamiento y un número n de días y halle el total de kilómetros acumulados hasta ese día (el n-ésimo día). Por ejemplo, si el primer día recorrió 5 Km....

  image.png

````
def montoya(km,n):
    r = km
    d = 2
    total = km

    while (d<=n):
        if (d%3 == 0):
            r = r / 2
        else:
            r = 2 * r + 10
        d = d + 1
        total = total + r

    return total

montoya(5, 10)
````
