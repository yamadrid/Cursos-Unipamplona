##Sentencia Def

La sentencia def es una definicin de funcin usada para crear objetos funciones definidas por el usuario.

La sintaxis para una definicin de funcin en Python es:

```
# def NOMBRE(LISTA_DE_PARAMETROS):
    """DOCSTRING_DE_FUNCION"""
    SENTENCIAS
    RETURN [EXPRESION]
```

A continuacin se detallan el significado de pseudo cdigo fuente anterior:



*   **NOMBRE**, es el nombre de la funcin.
*   **LISTA_DE_PARAMETROS**, es la lista de parmetros que puede recibir una funcin.
*   **DOCSTRING_DE_FUNCION**, es la cadena de caracteres usada para documentar la funcin.
*   **SENTENCIAS**, es el bloque de sentencias en cdigo fuente Python que realizar cierta operacin dada.
*   **RETURN**, es la sentencia return en cdigo Python.
*   **EXPRESION**, es la expresin o variable que devuelve la sentencia return.

````
def hola(arg):
  """El docstring de la funcin"""
  r = "Hola " + arg + "!"

  return r

hola("Mundo")
````

##Argumentos y parmetros



Al definir una funcin los valores los cuales se reciben se denominan parmetros, pero durante la llamada los valores que se envan se denominan argumentos.

###Por Posicin

Cuando envi argumentos a una funcin, estos se reciben por orden en los parmetros definidos. Se dice por tanto que son argumentos por posicin:

````
def resta(a, b):
  r = a - b
  return r

resta(30, 10)
````

###Por nombre

Sin embargo es posible evadir el orden de los parmetros si indica durante la llamada que valor tiene cada parmetro a partir de su nombre:

````
def resta(a, b):
  r = a - b
  return a - b

resta(b=30, a=10)
````

###Llamada sin argumentos

Al momento de llamar una funcin la cual tiene definidos unos parmetros, si no pasa los argumentos correctamente provocar una excepcin `TypeError`:

````
resta()
````
