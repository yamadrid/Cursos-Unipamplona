#Arreglos unidimensionales (vectores)

Un vector o arreglo unidimensional es una colección de datos del mismo tipo (en Python esta condición puede cambiar), referenciada con un solo nombre y que para acceder a cada uno de sus datos, se necesita un subíndice que en este caso indica la posición (de orden) donde se encuentra almacenado un dato, dentro de la colección.

image.png

En Python, estos arreglos estan divididos en varias notaciones particulares que llevan el mismo concepto con algunos cambios en su aplicación, para este caso se usara la notación de `Lista`. 

La notación para `Lista` es una secuencia de valores encerrados entre corchetes y separados por comas. Por ejemplo, si representamos a los alumnos mediante su código, se puede tener una lista de inscritos en la materia como la siguiente: 

$i = [78455, 89211, 66540, 45750]$. 

Al abrirse la inscripción, antes de que hubiera inscritos, la lista de inscritos se representará por una lista vacía: 

$i=[]$

```
i =[78455, 89211, 66540, 45750]

print(i)
```

## Métodos

El tipo de dato `lista` integra una serie de funciones y métodos integrados a continuación:

### Len()

Esta función obtiene el número de elementos de una lista

```
v = [2.5, 3.6, 4, 5]
print(len(v))
```

### Append()

Este método agrega un elemento al final de una lista

```
v = [2.5, 3.6, 4, 5]
print(v)
v.append(6)
print(v)
```

### Extend()

Este método extiende una lista agregando un iterable al final.

```
v = [2.1, 2.5, 3.6]
print(v)
v.extend([4])
print(v)
```

### Insert()

Este método inserta el elemento x en la lista, en el índice i.

```
v = [2.1, 2.5, 3.6, 4, 5, 6]
print(v)
v.insert(2, 3.7)
print(v)
```

### Pop()

Este método devuelve el último elemento de la lista, y lo borra de la misma.

```
v = [2.1, 2.5, 3.6, 4, 5, 6]
print(v.pop())
print(v)
```

### Remove()

Este método recibe como argumento un elemento, y borra su primera aparición en la lista.

```
v = [2.1, 2.5, 3.6, 4, 5, 6]
print(v)
v.remove(2.5)
print(v)
```

## Valores por Valor y Referencia

Dependiendo del tipo de dato que enviemos a la función, podemos diferenciar dos comportamientos:

**Paso por valor**: Se crea una copia local de la variable dentro de la función.

**Paso por referencia**: Se maneja directamente la variable, los cambios realizados dentro de la función le afectarán también fuera.

### Por valor

Como ya sabemos los números se pasan por valor y crean una copia dentro de la función, por eso no les afecta externamente lo que hagamos con ellos:

```
def doblar_valor(numero):
    numero *= 2

n = 10
doblar_valor(n)
print(n)
```

### Por referencia

Sin embargo las listas u otras colecciones, al ser tipos compuestos se pasan por referencia, y si se modifican dentro de la función se tara modificandola también fuera:

```
def doblar_valores(numeros):
    n = len(numeros)
    k = 0
    
    while k<n:
      numeros[k] *= 2
      k += 1

ns = [10,50,100]
doblar_valores(ns)
print(ns)
```
