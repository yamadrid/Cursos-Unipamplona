##Bucle While

El bucle while nos permite ejecutar un bloque de c√≥digo continuamente mientras la condici√≥n sea verdadera e incluso podemos crear bucles infinitos.

Estos se ejecutan mientras la condici√≥n sea verdadera, es decir, si la condici√≥n se cumple se ejecutara el cuerpo de dicho bucle y al finalizar se volver√° a comprobar la condici√≥n, si continua siendo verdadera pues se ejecutara nuevamente. 1. Hacer la funci√≥n Factorial

  $
  n!= \left\{ \begin{array}{lcc}
              0 &   si  & n = 0 \\
              \\ 1*2*3*4*....*n &  si & n > 0 \\
              \end{array}
    \right.
  $

````
def factorial(n):
    k = 1
    r = 1

    while (k<=n):
        r = r * k
        k = k + 1

    return r

factorial(5)
````

2. Dados dos (2) n√∫meros ‚Äùx‚Äù y ‚Äúz‚Äùdonde ‚Äúx‚Äù es menor que ‚Äúz‚Äù haga una funci√≥n para calcular la suma de los n√∫meros comprendidos entre ‚Äùx‚Äù y ‚Äúz‚Äù

````
def prob31(x,z):
    r = 1
    k = x + 1

    while (k<z):
        r = r * k
        k = k + 1

    return r

prob31(2,6)
````

3. Juan Pablo Montoya, entrena todos los d√≠as, recorriendo cierta cantidad de kil√≥metros, de tal forma que todos los d√≠as recorre el doble de lo que recorri√≥ el d√≠a anterior m√°s 10 km. Pero cada 3 d√≠as recorre solo la mitad de lo que recorri√≥ el d√≠a anterior. Realice una funci√≥n que dado el n√∫mero de kil√≥metros recorridos el primer d√≠a del entrenamiento y un n√∫mero n de d√≠as y halle el total de kil√≥metros acumulados hasta ese d√≠a (el n-√©simo d√≠a). Por ejemplo, si el primer d√≠a recorri√≥ 5 Km....

  image.png

````
def montoya(km,n):
    r = km
    d = 2
    total = km

    while (d<=n):
        if (d%3 == 0):
            r = r / 2
        else:
            r = 2 * r + 10
        d = d + 1
        total = total + r

    return total

montoya(5, 10)
````
